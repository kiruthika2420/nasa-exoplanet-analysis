```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# connect to NASA's database

```python
df=pd.read_csv("https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+pl_name,hostname,discoverymethod,disc_year,pl_rade,pl_bmasse,sy_dist+from+ps&format=csv")
print(df.head())
print(df.shape)
```

# cleaning of datas

```python
#cleaning null values
print(df.isnull().sum())
df_clean=df.dropna(subset=['pl_rade','sy_dist','disc_year'])
print(df_clean.shape)
#keeping only planets of radius 25 or lesser
max=df['pl_rade'].max()
print(max)
df_clean=df_clean[df_clean['pl_rade']<=25]
```

#grouping and aggregating

```python
discovery_year=df_clean.groupby('disc_year')
disc_pl_counts = discovery_year['pl_name'].count()
print(disc_pl_counts)
```

#visualisation

```python
sns.set_theme(style="darkgrid", rc={"axes.facecolor": "#11111e", "figure.facecolor": "#1b1b2f"})
scatter = plt.scatter(x=df_clean['sy_dist'], y=df_clean['pl_rade'], c=df_clean['disc_year'], cmap='plasma', alpha=0.7,edgecolors='none', s=25)
plt.title("Kepler & Beyond: Exoplanet Radius vs. Distance from Earth", fontsize=15, fontweight='bold', color='white')
plt.xlabel("Distance from Earth (Parsecs)", fontsize=12, color='white')
plt.ylabel("Planet Radius (Earth Radii = 1)", fontsize=12, color='white')

# Style the colorbar which tracks the discovery year
cbar = plt.colorbar(scatter)
cbar.set_label("Discovery Year", fontsize=12, color='white')
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='white')

# Styling
plt.xticks(color='white')
plt.yticks(color='white')

# Save
plt.savefig("exoplanet_analysis_chart.png", dpi=300, bbox_inches='tight', facecolor='#1b1b2f')
print("Chart successfully saved as 'exoplanet_analysis_chart.png'!")

plt.show()
```
